# Day 2

```python
strategy_guide = pd.read_csv('adventdata/day_2.csv', sep = ' ', names= ['opponent_move', 'my_move'])
```

```python
strategy_guide.head(8)
```

```python
scores_result = {
    'loose': 0,
    'win': 6,
    'draw': 3
}

scores_shape = {
    'Rock': 1,
    'Paper': 2,
    'Scissors': 3
}
```

```python
mapping_turn_opponent = {
    'A': 'Rock',
    'B': 'Paper',
    'C': 'Scissors'
}
```

```python
mapping_turn_myself = {
    'X': 'Rock',    
    'Y': 'Paper',
    'Z': 'Scissors'
}
```

```python
mapping_turn_result_what_should_it_be = {
    'X': 'loose',    
    'Y': 'draw',
    'Z': 'win'
}
```

```python
def define_move(move_of_opponent, expected_result):
    if expected_result == 'draw':
        return move_of_opponent
    
    if expected_result == 'win':
        if move_of_opponent == 'Rock':
            return 'Paper'
        if move_of_opponent == 'Paper':
            return 'Scissors'
        if move_of_opponent == 'Scissors':
            return 'Rock'
        
    if expected_result == 'loose':
        if move_of_opponent == 'Rock':
            return 'Scissors'
        if move_of_opponent == 'Paper':
            return 'Rock'
        if move_of_opponent == 'Scissors':
            return 'Paper'
```

```python
def win_or_loose(opponent_turn, my_turn):
    if opponent_turn == my_turn:
        return 'draw'
    if opponent_turn == 'Rock' and my_turn == 'Scissors':
        return 'loose'
    if opponent_turn == 'Scissors' and my_turn == 'Paper':
        return 'loose'
    if opponent_turn == 'Paper' and my_turn == 'Rock':
        return 'loose'
    return 'win'

```

```python
win_or_loose('Rock', 'Rock')
```

```python
# assignement 1
total_score = 0
for _, row in strategy_guide.iterrows():

    my_move = mapping_turn_myself[row['my_move']]
    opponent_move = mapping_turn_opponent[row['opponent_move']]

    result = win_or_loose(opponent_move, my_move)
    print(f"With my move {my_move} and opponent move {opponent_move}: {result}")

    score_of_result = scores_result[result]
    score_of_shape = scores_shape[my_move]

    total_score = total_score + score_of_result + score_of_shape
    
print(f"The total score is {total_score}")
```

```python
# assignement 2
total_score = 0
for _, row in strategy_guide.iterrows():

    my_outcome = mapping_turn_result_what_should_it_be[row['my_move']]
    opponent_move = mapping_turn_opponent[row['opponent_move']]

    my_move = define_move(opponent_move, my_outcome)

    result = win_or_loose(opponent_move, my_move)
    print(f"With my move {my_move} and opponent move {opponent_move}: {result}")

    score_of_result = scores_result[result]
    score_of_shape = scores_shape[my_move]

    total_score = total_score + score_of_result + score_of_shape
    
print(f"The total score is {total_score}")
```
